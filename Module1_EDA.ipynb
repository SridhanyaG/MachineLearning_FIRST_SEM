```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read data
train = pd.read_csv("../datathon23/data/train_data.csv")
test = pd.read_csv("../datathon23/data/test_data.csv")

# date
train["startdate"] = pd.to_datetime(train["startdate"])
test["startdate"] = pd.to_datetime(test["startdate"])
```

## 2. Descriptive analysis

Data overview https://www.kaggle.com/competitions/widsdatathon2023/data

```python
print(train.shape)
print(test.shape)

train.columns.difference(test.columns)
```

```python
for col in train.columns:
    print(col)
    
print(train.head())
```

```python
train.describe()
test.describe()
```

```python
# variable we want to predict on the test set
# contest-tmp2m-14d__tmp2m is the arithmetic mean of the max and min observed temperature over the next 14 days for each location and start date
target = "contest-tmp2m-14d__tmp2m"

plt.hist(train[target], bins = 20)
plt.title('Histogram of ' + target)

print(train[target].isna().sum())
train[target].describe()
```

```python
# climate regions using Koppen classification https://www.britannica.com/science/Koppen-climate-classification
train.climateregions__climateregion.value_counts()
```

```python
# number of unique locations in train set
lat_round = train.lat.multiply(100).round().astype(int)
lon_round = train.lon.multiply(100).round().astype(int)
train["lat_lon"] = "lat" + lat_round.astype(str) + "_lon" + lon_round.astype(str)
print(train.lat_lon.value_counts().unique())

# number of unique start date
print(train.startdate.value_counts().unique())

# number of unique location per climate region
region = 'climateregions__climateregion'
train.groupby(region)['lat_lon'].nunique().sort_values(ascending = False)
```

```python
# structure of test set
lat_round = test.lat.multiply(100).round().astype(int)
lon_round = test.lon.multiply(100).round().astype(int)
test["lat_lon"] = "lat" + lat_round.astype(str) + "_lon" + lon_round.astype(str)
print(test.lat_lon.value_counts())
print(test.startdate.value_counts())
```

```python
print("TRAIN SET\n min: " + str(min(train["startdate"])) + "\n max: " + str(max(train["startdate"])))
print("\n\nTEST SET\n min: " + str(min(test["startdate"])) + "\n max: " + str(max(test["startdate"])))
```

```python
train.groupby([region, 'lat_lon']).plot(x = 'startdate', y = target, ax = plt.gca(), legend = None)
```

```python
train["startday"] = train["startdate"].dt.strftime('%m/%d')
train["startyear"] = train["startdate"].dt.year

pivoted = train.pivot_table(index = ['lat_lon', 'startday'], columns = ['startyear'], values = target)
pivoted.columns = ['y' + str(col) for col in pivoted.columns.values]
pivoted.reset_index(inplace = True)

pivoted[pivoted['lat_lon'] == 'lat9_lon93'].set_index('startday').plot()
plt.title('lat9_lon93')
plt.ylabel(target)
```

## 4. Missing values 

Missing values can have a significant effect on the conclusions that can be drawn from the data. In addition, some models (e.g., logistic regression, support vector machines, neural networks) don't work with missing values. In that case, you need to impute the missing values.   

There are many methods to deal with missing values (dropping, imputation, predictive modeling). Here we are simply using the median as the imputed value. 

```python
#columns with NA's
na_cols = train.columns[train.isna().any()]

for c in na_cols:
    print(c + "   " + str(train[c].dtype))

# number of NA's per columns
train[na_cols].isna().sum()
train[na_cols].isna().sum() / train.shape[0] * 100 
```

```python
# imputation using the median
for c in na_cols:
    train[c].fillna(train[c].median(), inplace = True)
```

```python
# check that no more missing values
train.columns[train.isna().any()]
```
